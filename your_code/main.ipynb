#### Open the ` Employee.csv` file and name your dataset `employee`.

#### What's the mean salary by department?
Hint : There are 2 ways to do it - one way is with groupby and the other way is with pivot tables. Do it both ways. 


```python
# Method 1
employee.groupby(['Department']).agg({'Salary':'mean'})
```

```python
#Method 2
employee.pivot_table(index= ['Department'], values=['Salary'], aggfunc = ['mean'])
```

```python


#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title.
```

```python
employee.pivot_table(index= ['Department', 'Title'], values=['Salary'], aggfunc = ['mean']).reset_index()
```

#### We want to display the number of years of work experience and the salary for each Title. Add Years as a column in your pivot table.
*Hint: Use Pandas doc(see References in README.md) to explore the pivot_table() arguments*

```python
employee.pivot_table(values='Salary', index='Title', columns='Years', aggfunc=np.mean, fill_value=0) 
```

#### We want to know the information about the mean salary and number of employees in each department and for each job title.

```python
#employee.groupby(['Title','Department']).agg({'Salary':'mean', 'Name':'size'}).reset_index()
emp_dep_title= employee.pivot_table(index= ['Department','Title'], values = ['Salary', 'Name'], aggfunc = {'Salary':'mean','Name':'size'}).reset_index()
emp_dep_title.rename(columns={'Name': 'Number of Employees'}, inplace=True)
emp_dep_title
```

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
emp_stats= employee.pivot_table(index= ['Department','Title'], values = ['Salary'], aggfunc = {'Salary':['max','min','std','median']}).reset_index()
emp_stats
#the standard deviation for VP in HR and VP in IT and associate in IT  and associate in Sales is 0 because no difference in the salaries for each role in each department
#only analysts salaries varies :/
```

#### Based on your comments, fill the missing values with an appropriate value.

```python
emp_stats = emp_stats.fillna(0)
emp_stats
```

#### The stake holders want to know for each department the number of employees and how much money is spend on salaries. Could you provide that information? 

```python
stakeholder_view = employee.pivot_table(index= ['Department'], values = ['Salary', 'Name'], aggfunc = {'Salary':'sum','Name':'size'})
stakeholder_view.rename(columns={'Name': 'Number of Employees'}, inplace=True)
stakeholder_view
```

#### For each Department and Title, we want to know the the total years of work experience and the mean salary.

```python
employee.pivot_table(index= ['Department', 'Title'], values = ['Salary', 'Years'], aggfunc = {'Salary':'mean','Years':'sum'}).reset_index()
```
